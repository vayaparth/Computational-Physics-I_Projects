```python
#importing the useful libraries

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np
import math
```

- Without any air resistance, and only gravity acting upon the body, there would be a parabola shape.

- The parabola should be symmetric from the beginning to the end.

- the maximum height the body will be able to reach is when it's kinetic energy is zero, hence the velocity of the body will become 0 because of the gravity pulling it down and the body initially trying to move up.

- the maximum height can be calculated using the formula H = (0.5*((v_0)^2*sin(theta)^2) / g)


- with air resistance/ drag force, the shape of motion of the body won't be a simple parabola

- additionally the time of motion will be shorter due to the increased force going downwards which is gravity + drag force

Investigate:

The drag force of the body F^(f) = -k * m * |v|^n * v/|v|

```python

# Define constants
gravity = 9.81
drag_coefficient = 0.8
theta_deg = 34.0

#convert theta degrees to radians
theta_rad = math.radians(theta_deg)
```

```python

# Set the initial conditions
n= 1
height_x = 0
height_y = 0
velocity = 22.0
velocity_x = velocity * math.cos(theta_rad)
velocity_y = velocity * math.sin(theta_rad)
v_magnitude = math.sqrt(velocity_x**2 + velocity_y**2)
time_ground_without_k = (velocity_y/gravity) * 2 
#multiplied by 2 because it is the time the body reaches half way of the parabola since at that time the final velocity in y direvtion is zero.

print(velocity_x, "initial velocity in x direction")
print(velocity_y, "initial velocity in y direction")
print(time_ground_without_k,"- time it takes for object to fall on the ground")
print(v_magnitude, "this is the magnitude of the velocity")
```

Projectile Motion with friction and without

```python
# Euler-richardson algorithm
t = np.arange(0, 2.52, 0.01) 
# it starts at 0
# stops at 2.52 since the time it takes to fall would be 2.508
# each time step interval is 0.01 second

# usual projectile motion in vacuum - without friction
x = velocity_x*t 
y = velocity_y*t-0.5*(gravity)*t**2

# projectile motion with drag 
x1 = velocity_x*t-((drag_coefficient*velocity_x)*t**2)/2
y1 = velocity_y*t-0.5*(drag_coefficient*velocity_y+gravity)*t**2

# plotting 
plt.plot(x,y, label = "k = 0, only gravity")
plt.plot(x1, y1, label = "k = 0.8")
plt.grid()
plt.xlim(0, 50)
plt.ylim(0, 10)
```

C) now i will implement using n = 3/2 and n = 2 for medium-velocity friction and high-velocity friction

```python
n = 3/2
drag_coefficient1 = 0.05

#this is for no drag force
x = velocity_x*t 
y = velocity_y*t-0.5*(gravity)*t**2

#graph for the manually adjusted value for drga force
x2 = velocity_x*t-((drag_coefficient1*(v_magnitude**(n-1))*velocity_x)*t**2)/2
y2 = velocity_y*t-0.5*(drag_coefficient1*(v_magnitude**(n-1))*velocity_y+gravity)*t**2

#to calculate the exact drag_optimum like the previous problem
drag_optimum = drag_coefficient/math.sqrt(v_magnitude)
print(drag_optimum, "this is the exact k we should use to obtain the one where k=0.8 with power n = 1")

# projectile motion with optimum drag force to represent the previous motion
x3 = velocity_x*t-((drag_optimum*(v_magnitude**(n-1))*velocity_x)*t**2)/2
y3 = velocity_y*t-0.5*(drag_optimum*(v_magnitude**(n-1))*velocity_y+gravity)*t**2


# plotting 
plt.plot(x,y, label = "this is for the case with no gravity")
plt.plot(x2, y2, label = "graph to test out different values of k")
plt.plot(x3, y3, label = "this is for the when k is adjusted to the optimum value as in b)")

plt.grid()
plt.xlim(0, 50)
plt.ylim(0, 10)
```
